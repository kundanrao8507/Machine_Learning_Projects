```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
test_df = pd.read_excel('Test_set.xlsx')
```

```python
train_df.head()
```

```python
df = train_df.append(test_df)
```

```python
df.tail()
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df = df.drop(['Date_of_Journey'], axis = 1)
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minutes'] = df['Arrival_Time'].str.split(':').str[1]
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minutes'] = df['Arrival_Minutes'].astype(int)
df = df.drop(['Arrival_Time'], axis = 1)
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Minutes'] = df['Dep_Time'].str.split(':').str[1]
df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Minutes'] = df['Dep_Minutes'].astype(int)
df = df.drop(['Dep_Time'], axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isna()]
```

```python
df['Total_Stops'] =df['Total_Stops'].map({'non-stop': 0,'1 stop': 1,'2 stops': 2,'3 stops': 3,'4 stops': 4,'nan' : 1}) 
```

```python
df.head()
```

```python
df.drop(['Route'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df['Dh'] = df['Duration'].str.split(' ').str[0]
df['Dm'] = df['Duration'].str.split(' ').str[1]
df['Dh'] = df['Dh'].str[:-1]
df['Dm'] = df['Dm'].str[:-1]
```

```python
df.head()
```

```python
df['Dm'] = df['Dm'].fillna(0)
df['Duration'] = (df['Dh'].astype(int))*60+(df['Dm'].astype(int))
df.head()
```

```python
df.drop(['Dh'], axis = 1, inplace = True)
df.drop(['Dm'], axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df['Airline'] = labelencoder.fit_transform(df['Airline'])
df['Source'] = labelencoder.fit_transform(df['Source'])
df['Destination'] = labelencoder.fit_transform(df['Destination'])
df['Additional_Info'] = labelencoder.fit_transform(df['Additional_Info'])
```
